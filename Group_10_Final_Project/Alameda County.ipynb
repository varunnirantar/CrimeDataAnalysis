# Poisson distribution analysis for the Alameda County:
Now we shall fit a poisson distribution for the distribution of each crime type per month. We shall also build an interact to collect city, zip, and the number of crimes that may happen. Based on this, we shall show, current average number of crimes and the probabilty or chance that such number of crimes can happen. We shall use monthly average.
k - number of crimes that may happen in the next month L - lambda, mean number of crimes per month p - probability = (e^-L) * (L ^ k)/k!

```python
#Function for the interact
# If Zip = 'All', it will do the filtering based on Cities. Else, it will ignore City selection and filter by Zip code
def get_avg_crime(city,zip,crimedescr,k):
    if zip == 'All':
        #If City = 'All', then select all cities data else select only for  the required city
        if city == 'All':
            # Find monthly average by first forming a dataframe with the monthly counts grouped by CrimeDescription, year, and month
            # and then use SQL to get the monthly average for various crime types. This is for all cities and zip codes
            Crime_avg = sql.read_sql_query("""
            SELECT CrimeDescription, avg(Count) Average from(
            SELECT CrimeDescription,strftime('%Y',date) as Year, strftime('%m',date) as Month, count(*) as Count
            FROM alc_crimes
            where CrimeDescription not in ('None', '',' ')
            group by CrimeDescription, Year, Month)
            group by CrimeDescription
            order by 2 desc
            """, te)
        else:
            # This technique is similar to the one used above, but now we get the monthly average grouped by city. Finally, we 
            # filter the dataframe using the panda's filtering technique for the given city.
            Crime_avg = sql.read_sql_query("""
            SELECT CrimeDescription,City,avg(Count) Average from(
            SELECT CrimeDescription,City,strftime('%Y',date) as Year, strftime('%m',date) as Month, count(*) as Count
            FROM alc_crimes
            where CrimeDescription not in ('None', '',' ')
            group by CrimeDescription,City,Year, Month)
            group by CrimeDescription, City
            order by 3 desc, 2
            """, te)
            Crime_avg = Crime_avg[(Crime_avg['City'] == city)]
    else:
        # This technique is similar to the one used above, but now we get the monthly average grouped by zip. Finally, we 
            # filter the dataframe using the panda's filtering technique for the given zip.
        Crime_avg = sql.read_sql_query("""
        SELECT CrimeDescription,Zip,avg(Count) Average from(
           SELECT CrimeDescription,Zip,strftime('%Y',date) as Year, strftime('%m',date) as Month, count(*) as Count
            FROM alc_crimes
            where CrimeDescription not in ('None', '',' ')
            group by CrimeDescription,Zip,Year, Month)
            group by CrimeDescription, Zip
            order by 3 desc, 2
            """, te)
        zip_val = int(zip)
        Crime_avg = Crime_avg[(Crime_avg['Zip'] == zip_val)]            
   # Get the crime average for the criteria
    Crime_Average = Crime_avg[(Crime_avg['CrimeDescription'] == crimedescr)]
    L = Crime_Average['Average']
    if L.empty:
        L = 0.0
    else:
        L = float(L)
    print("Lambda, mean number of crimes per month: " + str("%.3f"%(L)) + "\n")
    # Now calculate the probability of k crimes happening in the next month
    p = np.exp(-L)*(L**k)/scipy.misc.factorial(k)
    print("Percentage probability of " + str(k) + " crimes happening: " + str("%.3f"%(p*100)) + " percent")
    #print("%.3f"%p)
    # Now calculate the probability that k or more crimes can happen
    pc = 0
    for n in range(k):
        p1 = np.exp(-L)*(L**n)/scipy.misc.factorial(n)
        #print("p[%d]=%f"%(n,p1))
        pc = p1 + pc
    pc = (1 - pc)
    print("\nPercentage probability of " + str(k) + " or more crimes happening: " + str("%.3f"%(pc*100)) + " percent")
    # Now let us plot Poisson distribution
    k1 = np.linspace(0,120,10)
    plt.plot(k1,np.exp(-L)*(L**k1)/scipy.misc.factorial(k1))
    plt.axis([0,120,0,0.1])
    plt.xlabel('Number of crimes that can happen in a month ',size=14)
    plt.ylabel('Probability',size=14)
    plt.title('Poisson plot',size=18)
    return(L)
```

#Poisson distribution:
Now we shall fit a poisson distribution for the distribution of homicides per month. We shall also build an interact to collect the number of homicides that may happen. Based on this, we shall show, current average number of homicides and the probabilty or chance that such number of homicides can happen. We shall use monthly average.

k - number of homicides that may happen in the next month
L - lambda, mean number of homicides per month
p - probability = (e^-L) * (L ^ k)/k!


```python
#Function for the interact
def get_avg_homicides(k):
    # Find monthly average of homicides
    homicides_avg = sql.read_sql_query("""
    SELECT round(avg(Count),0) as Average from(
    SELECT strftime('%Y',date) as Year, strftime('%m',date) as Month, count(*) as Count
    FROM opd_crimes
    where CType = "HOMICIDE"
    group by Year, Month)
    order by 1 desc
    """, te)          
    if homicides_avg.empty:
        L = 0.0
    else:
        L = float(homicides_avg['Average'])
    print("Lambda, mean number of homicides per month: " + str("%.3f"%(L)) + "\n")
    # Now calculate the probability of k homicides happening in the next month
    p = np.exp(-L)*(L**k)/scipy.misc.factorial(k)
    print("Percentage probability of " + str(k) + " homicides happening: " + str("%.3f"%(p*100)) + " percent")
    #print("%.3f"%p)
    # Now calculate the probability that k or more homicides can happen
    pc = 0
    for n in range(k):
        p1 = np.exp(-L)*(L**n)/scipy.misc.factorial(n)
        #print("p[%d]=%f"%(n,p1))
        pc = p1 + pc
    pc = (1 - pc)
    print("\nPercentage probability of " + str(k) + " or more homicides happening: " + str("%.3f"%(pc*100)) + " percent")
    # Now let us plot Poisson distribution
    k1 = np.linspace(0,120,10)
    plt.plot(k1,np.exp(-L)*(L**k1)/scipy.misc.factorial(k1))
    plt.axis([0,120,0,0.1])
    plt.xlabel('Number of homicides that can happen in a month ', size=18)
    plt.ylabel('Probability',size=14)
    plt.title('Poisson plot',size=14)
    return(L)
   
```

### Conclusion from Poisson analysis: Based on the historica data, there is more than 95% chance that 40 or more homicides will happen in Oakland in a month. There is about 52% chance of 50 or more homicides happening. However, there is only 1% chance of 70 or more homicides happening.
