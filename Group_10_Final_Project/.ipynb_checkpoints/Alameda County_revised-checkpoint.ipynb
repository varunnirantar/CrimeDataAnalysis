#Build an interact and folium
Based on the above dataset, build an interact with the following filters:
City, Zip and Year.

If Zip code is chosen, filtering on City will be ignored

Note: In order not to clutter the dataframe, choose only if the count is greater than 50

```python
#Function for the interact
# If Zip = 'All', it will do the filtering based on Cities. Else, it will ignore City selection and filter by Zip code
# Does size and color play any role here?
def plot_top_crimes(city, year, zip,color_val='r'):
    if zip == 'All':
        # Create a datafrme with CrimeDescription, City,and the count of the crimes
        Crime_Count = sql.read_sql_query("""
            SELECT CrimeDescription,City,strftime('%Y',date) as Year, count(*) as Count
            FROM alc_crimes
            where CrimeDescription not in ('None', '',' ') and City <> 'None'
            group by CrimeDescription,City,strftime('%Y',date) having count(*) > 50
            order by 1,2
            """, te)
        Crime_Count_Filtered = Crime_Count[(Crime_Count['City'] == city) & (Crime_Count['Year'] == year)]
    else:
        # Create a datafrme with CrimeDescription, Zip, and the count of the crimes
        Crime_Count = sql.read_sql_query("""
            SELECT CrimeDescription,Zip,strftime('%Y',date) as Year, count(*) as Count
            FROM alc_crimes
            where CrimeDescription not in ('None', '',' ') and City <> 'None'
            group by CrimeDescription,Zip,strftime('%Y',date) having count(*) > 50
            order by 1,2
            """, te)
        zip_val = int(zip)
        Crime_Count_Filtered = Crime_Count[(Crime_Count['Zip'] == zip_val) & (Crime_Count['Year'] == year)]
    ### Build a logic to stop plotting in case the count is equal to zero
    if Crime_Count_Filtered.empty:
        print("No data to plot")
    else:
        Crime_Count_Filtered.plot(kind='bar',x='CrimeDescription',y='Count',color=color_val)
        plt.title('Crime Type Visualization - Crimes that happened in a zip code/city')
        plt.xlabel('Crime_Type')
        plt.ylabel('Count')
```

#Build an interact to display monthly average for top 10 crimes in a given city or zip code

```python
#Function for the interact
# If Zip = 'All', it will do the filtering based on Cities. Else, it will ignore City selection and filter by Zip code
def plot_top_crimes(city,zip,size=10,color='red'):
    if zip == 'All':
        #If City = 'All', then select all cities data else select only for  the required city
        if city == 'All':
            # Find monthly average by first forming a dataframe with the monthly counts grouped by CrimeDescription, year, and month
            # and then use SQL to get the monthly average for various crime types. This is for all cities and zip codes
            Crime_avg = sql.read_sql_query("""
            SELECT CrimeDescription, avg(Count) Average from(
            SELECT CrimeDescription,strftime('%Y',date) as Year, strftime('%m',date) as Month, count(*) as Count
            FROM alc_crimes
            where CrimeDescription not in ('None', '',' ')
            group by CrimeDescription, Year, Month)
            group by CrimeDescription
            order by 2 desc
            """, te)
        else:
            # This technique is similar to the one used above, but now we get the monthly average grouped by city. Finally, we 
            # filter the dataframe using the panda's filtering technique for the given city.
            Crime_avg = sql.read_sql_query("""
            SELECT CrimeDescription,City,avg(Count) Average from(
            SELECT CrimeDescription,City,strftime('%Y',date) as Year, strftime('%m',date) as Month, count(*) as Count
            FROM alc_crimes
            where CrimeDescription not in ('None', '',' ')
            group by CrimeDescription,City,Year, Month)
            group by CrimeDescription, City
            order by 3 desc, 2
            """, te)
            Crime_avg = Crime_avg[(Crime_avg['City'] == city)]
    else:
        # This technique is similar to the one used above, but now we get the monthly average grouped by zip. Finally, we 
            # filter the dataframe using the panda's filtering technique for the given zip.
        Crime_avg = sql.read_sql_query("""
        SELECT CrimeDescription,Zip,avg(Count) Average from(
           SELECT CrimeDescription,Zip,strftime('%Y',date) as Year, strftime('%m',date) as Month, count(*) as Count
            FROM alc_crimes
            where CrimeDescription not in ('None', '',' ')
            group by CrimeDescription,Zip,Year, Month)
            group by CrimeDescription, Zip
            order by 3 desc, 2
            """, te)
        zip_val = int(zip)
        Crime_avg = Crime_avg[(Crime_avg['Zip'] == zip_val)]            
    ### Build a logic to stop plotting in case the count is equal to zero
    if Crime_avg.empty:
        print("No data to plot")
    else:
        Crime_avg = Crime_avg.head(10)
        Crime_avg.plot(kind='bar',x='CrimeDescription',y='Average')
        plt.title('Crime Type Visualization - Average number of crimes per month')
        plt.xlabel('Crime_Type')
        plt.ylabel('Average per month')
        plt.ylim((0,120))
```

# Poisson distribution:
Now we shall fit a poisson distribution for the distribution of each crime type per month. We shall also build an interact to collect city, zip, and the number of crimes that may happen. Based on this, we shall show, current average number of crimes and the probabilty or chance that such number of crimes can happen. We shall use monthly average.
k - number of crimes that may happen in the next month L - lambda, mean number of crimes per month p - probability = (e^-L) * (L ^ k)/k!

```python
#Function for the interact
# If Zip = 'All', it will do the filtering based on Cities. Else, it will ignore City selection and filter by Zip code
def get_avg_crime(city,zip,crimedescr,k):
    if zip == 'All':
        #If City = 'All', then select all cities data else select only for  the required city
        if city == 'All':
            # Find monthly average by first forming a dataframe with the monthly counts grouped by CrimeDescription, year, and month
            # and then use SQL to get the monthly average for various crime types. This is for all cities and zip codes
            Crime_avg = sql.read_sql_query("""
            SELECT CrimeDescription, avg(Count) Average from(
            SELECT CrimeDescription,strftime('%Y',date) as Year, strftime('%m',date) as Month, count(*) as Count
            FROM alc_crimes
            where CrimeDescription not in ('None', '',' ')
            group by CrimeDescription, Year, Month)
            group by CrimeDescription
            order by 2 desc
            """, te)
        else:
            # This technique is similar to the one used above, but now we get the monthly average grouped by city. Finally, we 
            # filter the dataframe using the panda's filtering technique for the given city.
            Crime_avg = sql.read_sql_query("""
            SELECT CrimeDescription,City,avg(Count) Average from(
            SELECT CrimeDescription,City,strftime('%Y',date) as Year, strftime('%m',date) as Month, count(*) as Count
            FROM alc_crimes
            where CrimeDescription not in ('None', '',' ')
            group by CrimeDescription,City,Year, Month)
            group by CrimeDescription, City
            order by 3 desc, 2
            """, te)
            Crime_avg = Crime_avg[(Crime_avg['City'] == city)]
    else:
        # This technique is similar to the one used above, but now we get the monthly average grouped by zip. Finally, we 
            # filter the dataframe using the panda's filtering technique for the given zip.
        Crime_avg = sql.read_sql_query("""
        SELECT CrimeDescription,Zip,avg(Count) Average from(
           SELECT CrimeDescription,Zip,strftime('%Y',date) as Year, strftime('%m',date) as Month, count(*) as Count
            FROM alc_crimes
            where CrimeDescription not in ('None', '',' ')
            group by CrimeDescription,Zip,Year, Month)
            group by CrimeDescription, Zip
            order by 3 desc, 2
            """, te)
        zip_val = int(zip)
        Crime_avg = Crime_avg[(Crime_avg['Zip'] == zip_val)]            
   # Get the crime average for the criteria
    Crime_Average = Crime_avg[(Crime_avg['CrimeDescription'] == crimedescr)]
    L = Crime_Average['Average']
    if L.empty:
        L = 0.0
    else:
        L = float(L)
    print("Lambda, mean number of crimes per month: " + str("%.3f"%(L)) + "\n")
    # Now calculate the probability of k crimes happening in the next month
    p = np.exp(-L)*(L**k)/scipy.misc.factorial(k)
    print("Percentage probability of " + str(k) + " crimes happening: " + str("%.3f"%(p*100)) + " percent")
    #print("%.3f"%p)
    # Now calculate the probability that k or more crimes can happen
    pc = 0
    for n in range(k):
        p1 = np.exp(-L)*(L**n)/scipy.misc.factorial(n)
        #print("p[%d]=%f"%(n,p1))
        pc = p1 + pc
    pc = (1 - pc)
    print("\nPercentage probability of " + str(k) + " or more crimes happening: " + str("%.3f"%(pc*100)) + " percent")
    # Now let us plot Poisson distribution
    k1 = np.linspace(0,120,10)
    plt.plot(k1,np.exp(-L)*(L**k1)/scipy.misc.factorial(k1))
    plt.axis([0,120,0,0.1])
    plt.xlabel('Number of crimes that can happen in a month ')
    plt.ylabel('Probability')
    plt.title('Poisson plot')
    return(L)
```

#Regression analysis of crime rates on home prices

```python
#Form a dataframe consisting of zip code, year & month in the format 'YYYY-MM', and count
Zip_Crime = sql.read_sql_query("""
            SELECT  Zip, strftime('%Y',date)||"-"||strftime('%m',date) as Year_Month, count(*) as Count
            FROM alc_crimes A
            where CrimeDescription not in ('None', '',' ')
            and Zip in (Select distinct RegionName from Zip_Median)
            and strftime('%Y',date) >= '2013'
            group by Zip, strftime('%Y',date), strftime('%m',date)
            order by 1,2
            """, te)
Zip_Crime
```

```python
Zip_Med = sql.read_sql_query("""
            Select RegionName as Zip, '2013-01' as Year_Month, [2013-01] as Median_Price from Zip_Median Union
            Select RegionName, '2013-02' as Year_Month, [2013-02] from Zip_Median Union
            Select RegionName, '2013-03' as Year_Month,[2013-03] from Zip_Median Union
            Select RegionName, '2013-04' as Year_Month,[2013-04] from Zip_Median Union
            Select RegionName, '2013-05' as Year_Month,[2013-05] from Zip_Median Union
            Select RegionName, '2013-06' as Year_Month,[2013-06] from Zip_Median Union
            Select RegionName, '2013-07' as Year_Month,[2013-07] from Zip_Median Union
            Select RegionName, '2013-08' as Year_Month,[2013-08] from Zip_Median Union
            Select RegionName, '2013-09' as Year_Month,[2013-09] from Zip_Median Union
            Select RegionName, '2013-10' as Year_Month,[2013-10] from Zip_Median Union
            Select RegionName, '2013-11' as Year_Month,[2013-11] from Zip_Median Union
            Select RegionName, '2013-12' as Year_Month,[2013-12] from Zip_Median Union
            Select RegionName, '2014-01' as Year_Month,[2014-01] from Zip_Median Union
            Select RegionName, '2014-02' as Year_Month,[2014-02] from Zip_Median Union
            Select RegionName, '2014-03' as Year_Month,[2014-03] from Zip_Median Union
            Select RegionName, '2014-04' as Year_Month,[2014-04] from Zip_Median Union
            Select RegionName, '2014-05' as Year_Month,[2014-05] from Zip_Median Union
            Select RegionName, '2014-06' as Year_Month,[2014-06] from Zip_Median Union
            Select RegionName, '2014-07' as Year_Month,[2014-07] from Zip_Median Union
            Select RegionName, '2014-08' as Year_Month,[2014-08] from Zip_Median Union
            Select RegionName, '2014-09' as Year_Month,[2014-09] from Zip_Median Union
            Select RegionName, '2014-10' as Year_Month,[2014-10] from Zip_Median
            order by 1,2
            """, te)
```

```python
Zip_Price_Crimes = pd.merge(Zip_Crime,Zip_Med)
#Zip_Price_Crimes.sort(columns=['Count'],ascending=False)
```

```python
#Function for the interact to perform home price regression analysis based on crime rate for a given zip code
def perform_reg_anlys(zip):
    zip_val = int(zip)
    Zip_Price_Crimes_1 = Zip_Price_Crimes[Zip_Price_Crimes['Zip']== zip_val]
    Zip_Price_Crimes_1.drop(['Zip','Year_Month'],axis=1)
    if Zip_Price_Crimes_1.empty:
        print ("No data to perform regression analysis")
    else:
        a = np.array(Zip_Price_Crimes_1['Count'])
        b = np.array(Zip_Price_Crimes_1['Median_Price'])
        slope, intercept, r_value, p_value, std_err = stats.linregress(a,b)
        print("Slope: " + str(slope))
        print("Intercept: " + str(intercept))
        print("r_value: " + str(r_value))
        print("p_value: " + str(p_value))
        print("R square: " + str(r_value*r_value))
        # Now let us plot the results of regression analysis
        Crimes = np.linspace(0,600,6)
        plt.plot(Crimes,intercept+(slope*Crimes))
        plt.axis([0,600,0,500000])
        plt.xlabel('Number of Crimes ')
        plt.ylabel('Median Home Price')
        plt.title('Regression plot for the selected Zip Code')
        # Now draw a scatter plot of the data: Median Price versus Crime Count
        Zip_Price_Crimes_1.plot(kind='scatter',x='Count',y='Median_Price')
        plt.title('Sctter - home price to crime count')
```
